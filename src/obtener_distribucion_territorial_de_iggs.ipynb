```python
path = f"./dataset/variables_alternas/DIRECCIONES_Y_TELEFONOS_MSPAS_1ER_SEMESTRE 2025.xlsx"
sheet_names = pd.ExcelFile(path).sheet_names
sheet_names
```

```python
dfs = [pd.read_excel(path, sheet_name=sheet, dtype=str) for sheet in sheet_names][:]
```

```python
columns = ["NOMBRE DEL SERVICIO *CATALOGADO COMO", "DISTRITO", "DIRECCIÓN FÍSICA", "No. DE TELÉFONO INSTITUCIONAL", "CORREO ELECTRÓNICO INSTITUCIONAL"]
for index, df in enumerate(dfs):
    df = df.drop(columns=df.columns[0])
    df.columns = columns
    df = df.dropna(how="all")
    df["depto"] = sheet_names[index]
    df["count_depto"] = df.shape[0]
    dfs[index] = df
    # print(df.info())
concat_df = pd.concat(dfs, axis=0)
```

```python
tipos_servicio = ['puesto de salud', 'centro de salud', 'puesto de salud', 'centro comunitario', 'centro de atención permanente', 'hospital', 'direccion', 'escuela', 'cementerio']
regex = r"(puesto de salud|centro de salud|centro comunitario|centro de atención permanente|hospital|direccion|escuela|cementerio)"
def get_tipo_servicio(nombre_servicio):
    match = re.search(regex, nombre_servicio.lower())
    if match:
        return match.group(0)
    return 'otro'
```

```python
concat_df['servicio'] = concat_df['NOMBRE DEL SERVICIO *CATALOGADO COMO'].apply(get_tipo_servicio)
concat_df['servicio']
```

```python
# Normalización de municipios y creación de lista de tuplas (municipio, departamento)
import unicodedata
def normalize_text(text):
    text = str(text).strip().lower()
    text = unicodedata.normalize('NFKD', text).encode('ASCII', 'ignore').decode('utf-8')
    return text
final_df["MUNICIPIO_NORM"] = final_df["DISTRITO"].apply(normalize_text)
final_df["DEPARTAMENTO_NORM"] = final_df["depto"].apply(normalize_text)
municipios_departamentos = list(zip(final_df["MUNICIPIO_NORM"], final_df["DEPARTAMENTO_NORM"]))
municipios_departamentos[:10]  # Muestra los primeros 10 para verificar
```

```python
# Supón que tienes una lista oficial de tuplas (municipio, departamento) llamada municipios_departamentos_oficial
# Ejemplo: municipios_departamentos_oficial = [("guatemala", "guatemala"), ("mixco", "guatemala"), ...]
from difflib import get_close_matches
def match_municipio(muni, muni_list):
    match = get_close_matches(muni, muni_list, n=1, cutoff=0.6)
    return match[0] if match else None

# Extrae la lista de municipios oficiales normalizados
municipios_oficiales = [normalize_text(m) for m, d in municipios_departamentos]
final_df["MUNICIPIO_NORM_MATCH"] = final_df["MUNICIPIO_NORM"].apply(lambda x: match_municipio(x, municipios_oficiales))
```

```python
final_df["MUNICIPIO_NORM_MATCH"] = final_df["DISTRITO"].apply(lambda x: match_municipio(x, municipios_oficiales))
```
