```python
from collections import defaultdict
from joblib import Parallel, delayed
from efficient_apriori import apriori as ea_apriori
from scipy.stats import chi2_contingency
from tqdm.auto import tqdm
import numpy as np
import pandas as pd
import math
import time
# --- CONFIG ---
n_jobs = -1
chi_pvalue_threshold = 0.05
require_expected_min = 5
min_pair_nonmissing = 50
min_support = 0.3
min_confidence = 0.6
min_lift = 1.0
min_conviction = 1.2
max_length = 2
# --------------
def chi2_test_pair(series_a, series_b):
    ct = pd.crosstab(series_a, series_b)
    if ct.shape[0] == 0 or ct.shape[1] == 0:
        return None
    try:
        chi2, p, dof, expected = chi2_contingency(ct, correction=False)
        return chi2, p, dof, expected
    except Exception:
        return None
def compute_all_chi2_tests(variables_enc, variables_orig_map, divisions_dict, years=range(2008,2024)):
    chi2_results = []

    pairs = [(variables_enc[i], variables_enc[j]) for i in range(len(variables_enc)) for j in range(i+1, len(variables_enc))]
    total_pairs = len(pairs)
    print(f"Total variable pairs to test: {total_pairs}")

    for division, year_map in divisions_dict.items():
        print(f"Processing division: {division}")
        if  isinstance(year_map, pd.DataFrame) and year_map.shape[0]==0:
            print(f"No data for division: {division}, skipping...")
            continue
        else:
            for year, (train_df, _) in year_map.items():
                df_train = train_df.copy()
                if df_train.shape[0] == 0:
                    continue

                for col_a_enc, col_b_enc in tqdm(pairs, desc=f"Year {year}", leave=False):
                    orig_col_a = variables_orig_map[col_a_enc]
                    orig_col_b = variables_orig_map[col_b_enc]

                    mask_valid = (df_train[orig_col_a] != 'ignorado') & (df_train[orig_col_b] != 'ignorado') & \
                                (df_train[orig_col_a] != 'Grupo ignorado') & (df_train[orig_col_b] != 'Grupo ignorado')
                    df_pair = df_train.loc[mask_valid, [col_a_enc, col_b_enc]].copy()

                    if df_pair.shape[0] < min_pair_nonmissing:
                        continue

                    chi_result = chi2_test_pair(df_pair[col_a_enc], df_pair[col_b_enc])
                    if chi_result is None:
                        continue

                    chi2, p, dof, expected = chi_result
                    chi2_results.append({
                        "division": division,
                        "year": year,
                        "col_a": orig_col_a,
                        "col_b": orig_col_b,
                        "chi2": chi2,
                        "p_value": p,
                        "dof": dof,
                        "expected_min": float(expected.min()),
                        "n_rows": df_pair.shape[0]
                    })

    return pd.DataFrame(chi2_results)
```

```python
from collections import defaultdict
from joblib import Parallel, delayed
from efficient_apriori import apriori as ea_apriori
from scipy.stats import chi2_contingency
from tqdm.auto import tqdm
import numpy as np
import pandas as pd
import math
import time

# --- CONFIG ---
n_jobs = -1
chi_pvalue_threshold = 0.05
require_expected_min = 5
min_pair_nonmissing = 50
min_support = 0.3
min_confidence = 0.6
min_lift = 1.0
min_conviction = 1.2
max_length = 2
# --------------

def chi2_test_pair(series_a, series_b):
    ct = pd.crosstab(series_a, series_b)
    if ct.shape[0] == 0 or ct.shape[1] == 0:
        return None
    try:
        chi2, p, dof, expected = chi2_contingency(ct, correction=False)
        return chi2, p, dof, expected
    except Exception:
        return None

def process_pair(df, col_a_enc, col_b_enc, orig_col_a, orig_col_b, year,
                 min_support=min_support, min_confidence=min_confidence):
    mask_valid = (df[orig_col_a] != 'ignorado') & (df[orig_col_b] != 'ignorado') & \
                 (df[orig_col_a] != 'Grupo ignorado') & (df[orig_col_b] != 'Grupo ignorado')
    df_pair = df.loc[mask_valid, [col_a_enc, col_b_enc]].copy()
    n_rows = len(df_pair)
    if n_rows < min_pair_nonmissing:
        return None

    chi_result = chi2_test_pair(df_pair[col_a_enc], df_pair[col_b_enc])
    if chi_result is None:
        return None
    chi2, p, dof, expected = chi_result
    metadata = (orig_col_a, orig_col_b, chi2, p, dof, expected.min(), n_rows, year)

    if (p >= chi_pvalue_threshold) or (expected.min() < require_expected_min):
        return None

    trans = df_pair.astype(str).apply(lambda r: [f"{col_a_enc}={r[col_a_enc]}", f"{col_b_enc}={r[col_b_enc]}"], axis=1).tolist()

    try:
        itemsets, rules = ea_apriori(trans, min_support=min_support, min_confidence=min_confidence, max_length=max_length)
    except Exception:
        return None

    if not rules:
        return None

    valid_rules = []
    for r in rules:
        if len(r.lhs) != 1 or len(r.rhs) != 1:
            continue
        if r.support >= min_support and r.confidence >= min_confidence and r.lift >= min_lift:
            conviction = getattr(r, 'conviction', None)
            if (conviction is None) or (conviction >= min_conviction):
                valid_rules.append(r)

    if not valid_rules:
        return None

    result = {
        "pair": (col_a_enc, col_b_enc),
        "orig_pair": (orig_col_a, orig_col_b),
        "n_rows": n_rows,
        "chi2": chi2,
        "p_value": p,
        "expected_min": float(expected.min()),
        "rules": valid_rules
    }
    return result, metadata

def find_correlated_pairs_and_rules(variables_enc, variables_orig_map, divisions_dict, years=range(2008,2024)):
    results = {}
    all_valid_pairs = []

    pairs = [(variables_enc[i], variables_enc[j]) for i in range(len(variables_enc)) for j in range(i+1, len(variables_enc))]
    total_pairs = len(pairs)
    print(f"Total variable pairs to consider: {total_pairs}")

    for division, year_map in divisions_dict.items():
        results[division] = {}
        print(f"Processing division: {division}")
        if isinstance(year_map, pd.DataFrame) and year_map.shape[0]==0:
            print(f"No data for division: {division}, skipping...")
            continue
        else:
            for year, (train_df, test_df) in year_map.items():
                df_train = train_df.copy()
                if df_train.shape[0] == 0:
                    results[division][year] = []
                    continue

                process = delayed(process_pair)
                parallel_results = Parallel(n_jobs=n_jobs)(
                    process(df_train, a, b, variables_orig_map[a], variables_orig_map[b], year) for (a, b) in pairs
                )

                pair_results = []
                metadata_list = []
                for r in parallel_results:
                    if r is not None:
                        result, metadata = r
                        pair_results.append(result)
                        metadata_list.append(metadata)

                results[division][year] = pair_results
                all_valid_pairs.extend(metadata_list)
                print(f"  Year {year}: {len(pair_results)} correlated pairs with valid rules found.")

    return results, all_valid_pairs

# Build the mapping from ENC columns to original
variables_enc = VARIABLES[:]  # your list with _ENC names
variables_orig_map = {v: v.replace('_ENC', '') for v in variables_enc}

# Run the optimized pipeline
t0 = time.time()
optimized_results, all_valid_pairs = find_correlated_pairs_and_rules(
    variables_enc=variables_enc,
    variables_orig_map=variables_orig_map,
    divisions_dict=sets,
    years=range(2008,2024)
)
t1 = time.time()
print("Total elapsed (s):", round(t1 - t0, 1))

# Optional: convert metadata to DataFrame
valid_pairs_df = pd.DataFrame(all_valid_pairs, columns=[
    "col_a", "col_b", "chi2", "p_value", "dof", "expected_min", "n_rows", "year"
])
```

```python
# ---------------------------
# FLATTEN optimized_results -> DataFrame (no changes to your pipeline)
# ---------------------------
def decode_item(item, label_encoders):
    """Decode one item like 'VIC_OCUP_ENC=12' -> 'VIC_OCUP=maestro' """
    col, val = item.split("=")
    val = int(val)
    orig_col = col.replace("_ENC", "")
    if orig_col in label_encoders:
        decoded_val = label_encoders[orig_col].inverse_transform([val])[0]
        return f"{orig_col}={decoded_val}"
    else:
        return item  # fallback
def explain_rules(results, label_encoders, max_rules=5):
    for division, years in results.items():
        print(f"\n=== Division: {division} ===")
        for year, pairs in years.items():
            if not pairs:
                continue
            print(f"\n Year {year}:")
            for pair_result in pairs:  # show first 3 pairs for brevity
                a, b = pair_result["orig_pair"]
                print(f"  Pair: {a} ↔ {b} (n={pair_result['n_rows']}, p={pair_result['p_value']:.4f})")
                for rule in pair_result["rules"]:
                    lhs = [decode_item(x, label_encoders) for x in rule.lhs]
                    rhs = [decode_item(x, label_encoders) for x in rule.rhs]
                    print(f"    Rule: {lhs} -> {rhs} "
                          f"(support={rule.support:.2f}, conf={rule.confidence:.2f}, lift={rule.lift:.2f}, conviction={rule.conviction:.2f})")

def _rule_to_row(pair_meta, division, year, rule, label_encoders):
    """Convert one Efficient-Apriori rule + pair metadata into a flat dict (one row)."""
    lhs = list(rule.lhs)
    rhs = list(rule.rhs)
    conviction = getattr(rule, "conviction", None)
    lhs_decoded = ";".join([decode_item(x, label_encoders) for x in rule.lhs])
    rhs_decoded = ";".join([decode_item(x, label_encoders) for x in rule.rhs])
    val_ant = lhs_decoded.split('=')[1]
    val_cns = rhs_decoded.split('=')[1]
    return {
        "DIVISION": division,
        "YEAR": year,
        "VARIABLE_ANTECEDENTE_ENC": pair_meta["pair"][0],
        "VARIABLE_CONSECUENTE_ENC": pair_meta["pair"][1],
        "VARIABLE_ANTECEDENTE": pair_meta["orig_pair"][0],
        "VARIABLE_CONSECUENTE": pair_meta["orig_pair"][1],
        "ANTECEDENT": lhs[0] if len(lhs) == 1 else ";".join(lhs),
        "CONSECUENT": rhs[0] if len(rhs) == 1 else ";".join(rhs),
        "VARIABLE_ANTECEDENTE_VALOR":val_ant,
        "VARIABLE_CONSECUENTE_VALOR":val_cns,
        "N_ROWS": int(pair_meta.get("n_rows", np.nan)),
        "CHI_2": float(pair_meta.get("chi2", np.nan)),
        "P_VALUE": float(pair_meta.get("p_value", np.nan)),
        "EXPECTED_MIN": float(pair_meta.get("expected_min", np.nan)),
        "SUPPORT": float(getattr(rule, "support", np.nan)),
        "CONFIDENCE": float(getattr(rule, "confidence", np.nan)),
        "LIFT": float(getattr(rule, "lift", np.nan)),
        "CONVICTION": float(conviction) if (conviction is not None) else np.nan,
    }

# Flatten structure
flat_rows = []
for division, year_map in optimized_results.items():
    # skip any non-year keys (guard)
    if division == "flat_rows":
        continue
    for year, pair_list in year_map.items():
        # pair_list is a list of pair_meta dicts returned by process_pair
        for pair_meta in pair_list:
            rules = pair_meta.get("rules", [])
            for r in rules:
                flat_rows.append(_rule_to_row(pair_meta, division, year, r, label_encoders))
# Valid pairs or not
valid_pairs_df.to_csv(r'./results_apriori/valores_validos_parejas.csv',index=False)
# Create DataFrame
df_apriori_rules = pd.DataFrame(flat_rows)

# Reorder columns for readability (skip missing)
cols_order = [
    "DIVISION","YEAR",
    "VARIABLE_ANTECEDENTE_ENC","VARIABLE_CONSECUENTE_ENC","VARIABLE_ANTECEDENTE","VARIABLE_CONSECUENTE",
    "ANTECEDENT","CONSECUENT", "VARIABLE_ANTECEDENTE_VALOR", "VARIABLE_CONSECUENTE_VALOR",
    "N_ROWS","CHI_2","P_VALUE","EXPECTED_MIN",
    "SUPPORT","CONFIDENCE","LIFT","CONVICTION"
]
cols = [c for c in cols_order if c in df_apriori_rules.columns] + [c for c in df_apriori_rules.columns if c not in cols_order]
df_apriori_rules = df_apriori_rules[cols]

# Save to CSV (single file)
out_path = "./results_apriori/apriori_filtered_rules_flat.csv"
df_apriori_rules.to_csv(out_path, index=False)
print(f"Saved {len(df_apriori_rules)} rule-rows to {out_path}")

# Optional: quick summary prints
if len(df_apriori_rules) > 0:
    print("Unique divisions:", df_apriori_rules['DIVISION'].nunique())
    print("Years range:", df_apriori_rules['YEAR'].min(), "-", df_apriori_rules['YEAR'].max())
    print("Top rules by support (head):")
    print(df_apriori_rules.sort_values("SUPPORT", ascending=False).head(10)[["YEAR","VARIABLE_ANTECEDENTE_ENC","VARIABLE_ANTECEDENTE","VARIABLE_CONSECUENTE","SUPPORT","CONFIDENCE","LIFT"]])
else:
    print("No rules found to flatten into DataFrame.")
```

```python
def parse_item(item):
    """Given 'COL=VALUE' return (col, value). If no '=', returns (item, None)."""
    if '=' in item:
        col, val = item.split('=', 1)
        return col, val
    return item, None

def compute_rule_chi2(pair_meta, rule, df_full, variables_orig_map):
    """
    pair_meta: dict returned by process_pair (contains orig_pair with original column names)
    rule: efficient-apriori Rule object (lhs, rhs)
    df_full: original dataframe used for that year/division (unfiltered)
    variables_orig_map: mapping encoded->orig (e.g., 'AGR_ALFAB_ENC' -> 'AGR_ALFAB')
    Returns dict with chi2 stats computed on the subset used for Apriori (ignorado-filtered).
    """
    # decode lhs/rhs items (expect single-element lists)
    lhs_item = list(rule.lhs)[0]
    rhs_item = list(rule.rhs)[0]
    lhs_col_enc, lhs_val_enc = parse_item(lhs_item)
    rhs_col_enc, rhs_val_enc = parse_item(rhs_item)

    # map encoded col names to original col names used to filter 'ignorado'
    orig_lhs_col = variables_orig_map.get(lhs_col_enc, lhs_col_enc).replace('_ENC','')
    orig_rhs_col = variables_orig_map.get(rhs_col_enc, rhs_col_enc).replace('_ENC','')

    # Build mask to remove 'ignorado' and 'Grupo ignorado' on both original columns
    mask_valid = (
        (df_full[orig_lhs_col] != 'ignorado') & (df_full[orig_lhs_col] != 'Grupo ignorado') &
        (df_full[orig_rhs_col] != 'ignorado') & (df_full[orig_rhs_col] != 'Grupo ignorado')
    )
    df_pair = df_full.loc[mask_valid, [lhs_col_enc, rhs_col_enc, orig_lhs_col, orig_rhs_col]].copy()

    # If either encoded column not in df_pair (rare), fall back to using original encoded columns
    if df_pair.shape[0] == 0:
        return {
            "rule_n": 0, "chi2": np.nan, "p_value": np.nan, "dof": np.nan,
            "expected_min": np.nan, "observed_matrix": None
        }

    # We need to compare category membership: build boolean series for antecedent and consequent
    # If lhs_val_enc is something like 'alfabeta' (string), we compare encoded column values stringified
    antecedent_mask = df_pair[lhs_col_enc].astype(str) == str(lhs_val_enc)
    consequent_mask = df_pair[rhs_col_enc].astype(str) == str(rhs_val_enc)

    # Contingency table: rows = antecedent True/False, cols = consequent True/False
    observed = pd.crosstab(antecedent_mask, consequent_mask)

    # If table has shape less than 2x2, pad with zeros to allow chi2_contingency
    # Ensure index [False, True] and columns [False, True]
    observed = observed.reindex(index=[False, True], columns=[False, True], fill_value=0)

    try:
        chi2, p, dof, expected = chi2_contingency(observed.values, correction=False)
        return {
            "rule_n": int(observed.values.sum()),
            "chi2": float(chi2),
            "p_value": float(p),
            "dof": int(dof),
            "expected_min": float(expected.min()),
            "observed_matrix": observed.values.tolist()
        }
    except Exception:
        return {
            "rule_n": int(observed.values.sum()),
            "chi2": np.nan,
            "p_value": np.nan,
            "dof": np.nan,
            "expected_min": np.nan,
            "observed_matrix": observed.values.tolist()
        }

# --- Integration: flatten optimized_results and compute chi2 per rule ---
flat_rows = []
# Provide original dataframes mapping so we can fetch df for each division/year.
# I assume `sets[division][year] = (train_df, test_df)` as earlier; use train_df used in pipeline.
for division, year_map in optimized_results.items():
    if division == "flat_rows":
        continue
    for year, pair_list in year_map.items():
        if not pair_list:
            continue
        # get the train_df you used for this division/year from sets
        try:
            train_df = sets[division][year][0]
        except Exception:
            train_df = None
        for pair_meta in pair_list:
            for r in pair_meta["rules"]:
                # base row (as before)
                row = {
                    "division": division,
                    "year": int(year),
                    "col_a_enc": pair_meta["pair"][0],
                    "col_b_enc": pair_meta["pair"][1],
                    "orig_col_a": pair_meta["orig_pair"][0],
                    "orig_col_b": pair_meta["orig_pair"][1],
                    "n_rows": int(pair_meta.get("n_rows", np.nan)),
                    "pair_chi2": float(pair_meta.get("chi2", np.nan)),
                    "pair_p_value": float(pair_meta.get("p_value", np.nan)),
                    "pair_expected_min": float(pair_meta.get("expected_min", np.nan)),
                    "support": float(getattr(r, "support", np.nan)),
                    "confidence": float(getattr(r, "confidence", np.nan)),
                    "lift": float(getattr(r, "lift", np.nan)),
                    "conviction": float(getattr(r, "conviction", np.nan)) if getattr(r, "conviction", None) is not None else np.nan,
                    "antecedent_enc": list(r.lhs)[0],
                    "consequent_enc": list(r.rhs)[0]
                }

                # compute chi2 between antecedent and consequent using original train df
                if train_df is not None:
                    chi2_stats = compute_rule_chi2(pair_meta, r, train_df, variables_orig_map)
                    row.update({
                        "rule_n": chi2_stats["rule_n"],
                        "rule_chi2": chi2_stats["chi2"],
                        "rule_p_value": chi2_stats["p_value"],
                        "rule_dof": chi2_stats["dof"],
                        "rule_expected_min": chi2_stats["expected_min"],
                        "rule_observed": chi2_stats["observed_matrix"]
                    })
                else:
                    row.update({
                        "rule_n": np.nan,
                        "rule_chi2": np.nan,
                        "rule_p_value": np.nan,
                        "rule_dof": np.nan,
                        "rule_expected_min": np.nan,
                        "rule_observed": None
                    })

                flat_rows.append(row)

# build df
df_rules_with_rule_chi2 = pd.DataFrame(flat_rows)

# reorder and save
cols = [
    "division","year","orig_col_a","orig_col_b","col_a_enc","col_b_enc","n_rows",
    "pair_chi2","pair_p_value","pair_expected_min",
    "antecedent_enc","consequent_enc","support","confidence","lift","conviction",
    "rule_n","rule_chi2","rule_p_value","rule_dof","rule_expected_min","rule_observed"
]
cols = [c for c in cols if c in df_rules_with_rule_chi2.columns] + [c for c in df_rules_with_rule_chi2.columns if c not in cols]
df_rules_with_rule_chi2 = df_rules_with_rule_chi2[cols]

df_rules_with_rule_chi2.to_csv("./results_apriori/apriori_rules_with_rule_chi2.csv", index=False, encoding="utf-8-sig")
print("Saved apriori_rules_with_rule_chi2.csv with", len(df_rules_with_rule_chi2), "rows")
```

```python
def collect_rules_to_df(results, label_encoders):
    """
    Flatten results into a pandas DataFrame with decoded antecedents/consequents and metrics.
    """
    rows = []
    for division, years in results.items():
        for year, pairs in years.items():
            if not pairs:
                continue
            for pair_result in pairs:
                for rule in pair_result["rules"]:
                    lhs = [decode_item(x, label_encoders) for x in rule.lhs]
                    rhs = [decode_item(x, label_encoders) for x in rule.rhs]
                    rows.append({
                        "DIVISION": division,
                        "YEAR": year,
                        "VARIABLE_ANTECEDENTE": ";".join(lhs),
                        "VARIABLE_CONSECUENTE": ";".join(rhs),
                        "SUPPORT": round(rule.support, 4),
                        "CONFIDENCE": round(rule.confidence, 4),
                        "LIFT": round(rule.lift, 4),
                        "CONVICTION": round(getattr(rule, "conviction", np.nan), 4)
                    })
    df = pd.DataFrame(rows)
    df[['VARIABLE_ANTECEDENTE', 'ANTECEDENTE_VALOR']] = df['VARIABLE_ANTECEDENTE'].str.split('=', n=1, expand=True)
    df[['VARIABLE_CONSECUENTE', 'CONSECUENTE_VALOR']] = df['VARIABLE_CONSECUENTE'].str.split('=', n=1, expand=True)
    return df

# Generate DataFrame
rules_df = collect_rules_to_df(optimized_results, label_encoders)

# Save to CSV and Excel
rules_df.to_csv("./results_apriori/association_rules.csv", index=False, encoding="utf-8-sig")
rules_df.to_excel("./results_apriori/association_rules.xlsx", index=False)

print("Saved association_rules.csv and association_rules.xlsx")
rules_df.head()
```
