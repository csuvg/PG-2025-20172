## Reemplazo de valores nulos

### Generar función para chequear si hay nulos en la columna

```python
def has_nulls(df, column):
    return df[column].isnull().sum()
```

#### Rellenar a qué se dedica la víctima (no ingresos)

```python
df['VIC_DEDICA'] = df['VIC_DEDICA'].fillna('Ignorado')
df['VIC_DEDICA'].value_counts()
```

#### Rellenar tipo de discapacidad de la victima
- De manera que en el caso de que no tenga discapacidad entonces el tipo debe ser no aplica o sin discapacidad.
- Si se ignora que tenga o no una discapacidad entonces debe de ser ignorado el tipo de discapacidad ya que no se sabe si tiene no.
- Si se sabe que tiene una discapacidad pero se desconoce el tipo entonces se pone ignorado ya que se ignora el tipo de discapacidad


```python
df.loc[df["VIC_DISC"] == "Sin discapacidad", "TIPO_DISCAQ"] = df.loc[df["VIC_DISC"] == "Sin discapacidad", "TIPO_DISCAQ"].fillna("Sin discapacidad")
df.loc[df["VIC_DISC"] == "Ignorado", "TIPO_DISCAQ"] = df.loc[df["VIC_DISC"] == "Ignorado", "TIPO_DISCAQ"].fillna("Ignorado")
df.loc[df["VIC_DISC"] == "Con discapacidad", "TIPO_DISCAQ"] = df.loc[df["VIC_DISC"] == "Con discapacidad", "TIPO_DISCAQ"].fillna("Tipo de discapacidad no especificado")
df['TIPO_DISCAQ'].value_counts()
```
